#Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
print(df.head())
```

#Change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5-10", "5-oct")
df.head()
```

Gender distribution

```python
plt.figure(figure = (5,4))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

#Above shows female data is more than male

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figure = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship b/w parent's education and student's scores")
plt.show()
```

```python
# Counting the number of students in each ethnic group
groupA = df.loc[df['EthnicGroup'] == "group A"].shape[0]
groupB = df.loc[df['EthnicGroup'] == "group B"].shape[0]
groupC = df.loc[df['EthnicGroup'] == "group C"].shape[0]
groupD = df.loc[df['EthnicGroup'] == "group D"].shape[0]
groupE = df.loc[df['EthnicGroup'] == "group E"].shape[0]

# List of groups and their respective counts
l = ['group A', 'group B', 'group C', 'group D', 'group E']
mlist = [groupA, groupB, groupC, groupD, groupE]

# Plotting the pie chart
plt.pie(mlist, labels=l, autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
